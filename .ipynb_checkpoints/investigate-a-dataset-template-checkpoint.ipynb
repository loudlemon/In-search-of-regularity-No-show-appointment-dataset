"no-shows" \

Before we proceed, let's make a few transformations with some variables, and than take a look at heatmap \
and grasp some relations between variables:
 - Column name 'No-show' seems a bit confusing, so let's rename it to 'show_up' and (be careful!) change 'No'\
   to 1,meaning a patient showed up and 'Yes' to 0 when a patient didn't show oneself and then replace 1's with 'Yes and 0's with 'No' to avoid confusing
 - Drop 'AppointmentID' and 'Patientid' columns,because they represent just a unique number of appointment/patient, doesn't  seem
   to be participating in "No-show' pattern
 - rename some columns with typos
 - Convert 'ScheduledDay' and 'AppointmentDay' columns to 'datetime'
 
That's all for now, let's see what we can observe from the data

```python
df.rename({'No-show':'showed_up','Hipertension':'Hypertension','Handcap':'Handicap'},axis = 1,inplace = True)
df['showed_up'].replace({'No':1,'Yes':0},inplace = True)
df['showed_up'].replace({1:'Yes',0:'No'},inplace = True)
```

```python
df.drop(['AppointmentID','PatientId'],axis = 1,inplace = True)
```

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'],yearfirst = True).dt.date.astype('datetime64[ns]')
```

```python
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'],yearfirst = True).dt.date.astype('datetime64[ns]')
```

```python
df.head(4)
```

For future plots let's write a small and simple function


```python
def plotter(data_x = None,data_hue = None,dataframe = None,kind = None,print_number = None):
    
    """ Just a function to plot some bars and distributions"""
    
    if kind == 'countplot':
        ax = sns.countplot(x=data_x, hue=data_hue, data=dataframe)
        ax.set_title("Shows and no-shows for Females and Males")
        labels=['Female', 'Male']
        ax.set_xticklabels(labels)
        plt.show()
        M = df[(df['Gender'] == 'M') & (df['showed_up'] == 'Yes')].shape[0] # count the number of 'showed_up'
        F = df[(df['Gender'] == 'F') & (df['showed_up'] == 'Yes')].shape[0] # the same for women
        if print_number == True:
            print('Number of women showed up is {}\nNumber of men showed up is {}'.format(F,M))
        
    elif kind == 'distplot':
        M = df[df['Gender'] == 'M']
        F = df[df['Gender'] == 'F']
        fig,ax = plt.subplots(3,figsize = (10,8))
        sns.distplot(df['Age'],color = 'g',ax = ax[0])
        ax[0].set_title('General distribution of Age variable')
        ax[1].set_title('Distribution of Men\'s age')
        ax[2].set_title('Distribution of Women\'s age')
        sns.distplot(M['Age'],color = 'b',ax = ax[1])
        sns.distplot(F['Age'],color = 'm',ax = ax[2])
        plt.tight_layout()
        
       
```

```python
plotter(df['Gender'],df['showed_up'],df,kind = 'countplot')
```

So,our first assumption was correct, and amongst those who visited there are more women than men,but we should not follow this idea because our represantation is not balanced

### Answering questions

Let's write a function for more complicated plots and to answer the question:\
**Is it useful to send sms with a reminder about appointment date?**

```python
def plot_agg(column_to_show,):
    
    """ Here we will try to understand relations between attendance,sms service and deseases
    in different neigbourhoods"""
    
    fig, (ax1,ax2,ax3,ax4,ax5) = plt.subplots(5,figsize=(10,14))
    
    if column_to_show == 'Scholarship':
        ax1.set(title = 'Attendance of participants of "Bolsa Familia" programm in ' + 
                'JARDIM CAMBURI neighbourhood')
        ax2.set(title = 'Attendance of participants of "Bolsa Familia" programm in ' + 
                'MARIA ORTIZ neighbourhood')
        ax3.set(title = 'Attendance of participants of "Bolsa Familia" programm in ' +
                'RESISTÊNCIA neighbourhood')
        ax4.set(title = 'Attendance of participants of "Bolsa Familia" programm in ' +
                'JARDIM DA PENHA neighbourhood')
        ax5.set(title = 'Attendance of participants of "Bolsa Familia" programm in ' +
                'ITARARÉ neighbourhood')
    elif column_to_show == 'Hypertension':
        ax1.set(title = 'Attendance of participants with HYPERTENSION in ' + 
                'JARDIM CAMBURI neighbourhood')
        ax2.set(title = 'Attendance of participants with HYPERTENSION in ' + 
                'MARIA ORTIZ neighbourhood')
        ax3.set(title = 'Attendance of participants with HYPERTENSION in ' +
                'RESISTÊNCIA neighbourhood')
        ax4.set(title = 'Attendance of participants with HYPERTENSION in ' +
                'JARDIM DA PENHA neighbourhood')
        ax5.set(title = 'Attendance of participants with HYPERTENSION in ' +
                'ITARARÉ neighbourhood')
        
    elif column_to_show == 'Diabetes':
        ax1.set(title = 'Attendance of participants with DIABETES in ' + 
                'JARDIM CAMBURI neighbourhood')
        ax2.set(title = 'Attendance of participants with DIABETES in ' + 
                'MARIA ORTIZ neighbourhood')
        ax3.set(title = 'Attendance of participants with DIABETES in ' +
                'RESISTÊNCIA neighbourhood')
        ax4.set(title = 'Attendance of participants with DIABETES in ' +
                'JARDIM DA PENHA neighbourhood')
        ax5.set(title = 'Attendance of participants with DIABETES in ' +
                'ITARARÉ neighbourhood')
        
    elif column_to_show == 'Alcoholism':
        ax1.set(title = 'Attendance of participants with ALCOHOL addiction in ' + 
                'JARDIM CAMBURI neighbourhood')
        ax2.set(title = 'Attendance of participants with ALCOHOL addiction in ' + 
                'MARIA ORTIZ neighbourhood')
        ax3.set(title = 'Attendance of participants with ALCOHOL addiction in ' +
                'RESISTÊNCIA neighbourhood')
        ax4.set(title = 'Attendance of participants with ALCOHOL addiction in ' +
                'JARDIM DA PENHA neighbourhood')
        ax5.set(title = 'Attendance of participants with ALCOHOL addiction in ' +
                'ITARARÉ neighbourhood')
         
    df_jardim.groupby(['SMS_received','showed_up'])[column_to_show].count().unstack().plot(
            ax=ax1,kind = 'bar')
    df_maria.groupby(['SMS_received','showed_up']).count()[column_to_show].unstack().plot(
            ax=ax2,kind = 'bar')
    df_resist.groupby(['SMS_received','showed_up']).count()[column_to_show].unstack().plot(
            ax=ax3,kind = 'bar')
    df_jardim_p.groupby(['SMS_received','showed_up']).count()[column_to_show].unstack().plot(
            ax=ax4,kind = 'bar')
    df_itarare.groupby(['SMS_received','showed_up']).count()[column_to_show].unstack().plot(
            ax=ax5,kind = 'bar')
    plt.tight_layout()
                
    
    
                
```


plot bars with different neighbourhoods and deseases

```python
cols = ['Scholarship','Hypertension','Diabetes','Alcoholism']
for des in cols:
    plot_agg(des)
```

As from what we can see our sms service does not have any influence om people from different areas with different ailments,so the answer is **No, it doesn't make people more conscious about their scheduled appointment**\
And what about 'Bolsa Familia' programm? Does it have some influence on people's mind?
